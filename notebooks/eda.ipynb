## 🔍 Explore the Data

## Import Libraries

```python
import pandas as pd
```

## Load All WHO Excel Files

```python
files = {
    "coverage": "../data/raw/coverage-data.xlsx",
    "incidence": "../data/raw/incidence-rate-data.xlsx",
    "cases": "../data/raw/reported-cases-data.xlsx",
    "intro": "../data/raw/vaccine-introduction-data.xlsx",
    "schedule": "../data/raw/vaccine-schedule-data.xlsx"
}

dfs = {}

for name, path in files.items():
    dfs[name] = pd.read_excel(path)

print("All datasets loaded successfully!")
```

## Check Dataset Shape

```python
for name, df in dfs.items():
    print("="*60)
    print(f"{name.upper()} DATASET")
    print("="*60)
    print("Shape:", df.shape)
```

## View First Five Rows

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.head())
```

## View Column Names

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.columns.tolist())
```

## Check Data Types

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.dtypes)
```

## Check Missing Values

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.isnull().sum())
```

## To see missing values as percentages:

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    missing = (df.isnull().sum() / len(df)) * 100
    print(missing.sort_values(ascending=False))
```

## Check Duplicate Rows

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print("Duplicate Rows:", df.duplicated().sum())
```

## Summary Statistics

## For numerical columns:

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.describe())
```

## For all columns (including text):

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    print(df.describe(include="all"))
```

## Dataset Information

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())
    df.info()
```

## Unique Values (Optional)

```python
for name, df in dfs.items():
    print("="*60)
    print(name.upper())

    for col in df.columns:
        print(f"{col}: {df[col].nunique()} unique values")
```

## Generate a Complete Data Quality Report

## Instead of running many cells, you can generate a concise report for each dataset:

```python
for name, df in dfs.items():
    print("\n" + "="*70)
    print(f"DATASET: {name.upper()}")
    print("="*70)

    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicate Rows:", df.duplicated().sum())

    print("\nSummary Statistics:")
    print(df.describe(include="all"))
```
